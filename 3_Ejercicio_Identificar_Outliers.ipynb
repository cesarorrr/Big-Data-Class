```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Creamos un DataFrame de ejemplo con datos aleatorios y algunos outliers
# Crear datos aleatorios
data = {
    'A': np.random.normal(loc=0, scale=1, size=1000),  # Normal con media 0 y desviación estándar 1
    'B': np.random.normal(loc=0, scale=1, size=1000),  # Normal con media 0 y desviación estándar 1
    'C': np.random.normal(loc=0, scale=1, size=1000),  # Normal con media 0 y desviación estándar 1
    'D': np.random.normal(loc=10, scale=5, size=1000)   # Normal con media 10 y desviación estándar 5
}
df = pd.DataFrame(data)
```

```python
 print(df)
```

```python
# Visualización de los datos utilizando pairplot de Seaborn
sns.pairplot(df)
plt.show()
```

```python
# Visualización de los boxplots para detectar outliers
sns.boxplot(df)
plt.show()
```
