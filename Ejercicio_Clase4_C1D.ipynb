Creamos la tabla y subimos los datos

```python
# Crear la tabla (si no existe)
cursor.execute('''
    CREATE TABLE IF NOT EXISTS teams (
        id INTEGER PRIMARY KEY,
        full_name TEXT NOT NULL,
        abbreviation TEXT NOT NULL,
        nickname TEXT NOT NULL,
        city TEXT NOT NULL,
        state TEXT NOT NULL,
        year_founded REAL
    )
''')

# Insertar los datos del DataFrame en la tabla
for index, row in teams.iterrows():
    cursor.execute('''
        INSERT INTO teams (id, full_name, abbreviation, nickname, city, state, year_founded)
        VALUES (?, ?, ?, ?, ?, ?, ?)
    ''', (row['id'], row['full_name'], row['abbreviation'], row['nickname'], row['city'], row['state'], row['year_founded']))

# Guardar los cambios
conn.commit()
```
