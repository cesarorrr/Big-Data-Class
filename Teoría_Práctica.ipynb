```python
# Importamos Libreria
import seaborn as sns
import matplotlib.pyplot as plt

# Hacemos un histograma de ventas
sns.histplot(ventas["Sales"])
plt.title('Sales Count')
```

```python
# Hacemos un histograma de fechas
sns.histplot(ventas["Date"])
plt.title('Date Count')
```

```python
# Importamos libreria
import plotly.graph_objects as go
# Creamos la figura
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=ventas["Date"],
    y=ventas["Sales"],
    mode="markers",
    name="Sales Scatter",
    marker=dict(color="red")
))

fig.add_trace(go.Scatter(
    x=ventas["Date"],
    y=ventas["Sales"],
    mode="markers",
    name="Sales Scatter",
    marker=dict(color="red")
))


```

```python
# Calcular media
mean = ventas["Sales"].mean()
# Calcular mediana
median = ventas["Sales"].median()
#Calcular moda
mode = ventas["Sales"].mode()[0]

print(f"Mean:{round(mean,2)}")
print(f"Median:{round(median,2)}")
print(f"Mode:{round(mode,2)}")
```

```python
# Importamos numpy
import numpy as np

ventas["Sales Detector"] = np.where(ventas["Sales"]>25000,"High","Low")
ventas.head()
```

```python
# Extraer año de la fecha
# Usamos el dt para extraer informacion de un formato date

ventas["Year"] = ventas["Date"].dt.year
ventas.head()
```

```python
# Eliminar una columna y crear un nuevo dataframe sin esta columna
# axis=0:Eliminan filas
# axis=1:Eliminan columnas
ventas_wdate = ventas.drop("Date",axis=1)

print(ventas)
print(ventas_wdate)

# ventas.head()
# ventas_wdate.head()
```
